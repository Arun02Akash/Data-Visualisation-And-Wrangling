```python
#Random Over-Sampling

# class count
class_count_0, class_count_1 = data['Outcome'].value_counts()

# Separate class
class_0 = data[data['Outcome'] == 0]
class_1 = data[data['Outcome'] == 1]
# print the shape of the class
print('Outcome 0:', class_0.shape)
print('Outcome 1:', class_1.shape)
```

```python
plt.figure(figsize=(10,5))
pd.value_counts(data['Outcome']).sort_index().plot(kind='bar',color=['orange','lightgreen'])
plt.title('Diabetes data')
plt.xlabel('Presence of Diabetes')
plt.ylabel('No of patients')
plt.xticks([0,1],['No Diabetes','Diabetes'],rotation=0)
plt.show()
```

```python
class_1_over = class_1.sample(class_count_0, replace=True)
test_over = pd.concat([class_1_over, class_0], axis=0)
print("total class of 1 and 0:",test_over['Outcome'].value_counts())# plot the count after under-sampeling
plt.figure(figsize=(10,5))
pd.value_counts(test_over['Outcome']).sort_index().plot(kind='bar', title='count (target)',color=['orange','lightgreen'])
plt.xticks([0,1],['No Diabetes','Diabetes'],rotation=0)
plt.xlabel('Presence of Diabetes')
plt.ylabel('No of patients')
plt.show()
```

```python
X=data[['Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']]
```

```python
X.shape
```

```python
y=data['Outcome']
```

```python
y.shape
```

```python
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2147, random_state=45)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression(random_state=0,max_iter=3000)
LR.fit(X_train, y_train)
p1=LR.score(X_test,y_test)*100
print(p1)
```

```python
from sklearn.ensemble import AdaBoostClassifier
ADA=AdaBoostClassifier()
ADA.fit(X_train, y_train)
p2=ADA.score(X_test,y_test)*100
print(p2)
```

```python
from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier(max_features='auto', n_estimators=200)
RF.fit(X_train, y_train)
p3=RF.score(X_test,y_test)*100
print(p3)
```

```python
from sklearn.tree import DecisionTreeClassifier
DC=DecisionTreeClassifier()
DC.fit(X_train, y_train)
p4=DC.score(X_test,y_test)*100
print(p4)
```

```python
from sklearn.naive_bayes import GaussianNB
GB=GaussianNB()
GB.fit(X_train,y_train)
p5=GB.score(X_test,y_test)*100
print(p5)
```

```python
a=["LogisticRegression","AdaBoostClassifier","RandomForestClassifier","DecisionTreeClassifier","GaussianNB"]
b=[p1,p2,p3,p4,p5]
plt.figure(figsize=(16,6))
plt.bar(a,b,width=0.5,color='blue')
plt.title("Accuracy Graph")
plt.xlabel("Algorithm")
plt.ylabel("percentage")
plt.show()
```

```python
import pickle
with open("dia_model.pickle","wb") as g:
    pickle.dump(RF,g,protocol=2)
```

```python
actual=y_test
predicted=RF.predict(X_test)
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
matrix = confusion_matrix(actual,predicted, labels=[0,1])
print('Confusion matrix : \n',matrix)
```

```python
# outcome values order in sklearn
tp, fn, fp, tn = confusion_matrix(actual,predicted,labels=[1,0]).reshape(-1)
print('Outcome values : \n', tp, fn, fp, tn)
print("True positive : ",tp)
print("False negative: ",fn)
print("False positive: ",fp)
print("True negative : ",tn)
```

```python
# classification report for precision, recall f1-score and accuracy
from sklearn.metrics import classification_report
matrix = classification_report(actual,predicted,labels=[1,0])
print('Classification report : \n',matrix)
```

```python
import seaborn as sns
output_labels=['No Diabetes','Diabetes']
cm = confusion_matrix(actual,predicted)
#print(cm)
cm=pd.DataFrame(cm,index=['0','1'],columns=['0','1'])
plt.figure(figsize=(12,6))
sns.heatmap(cm,cmap='Blues',annot=True,fmt='',xticklabels=output_labels,yticklabels=output_labels)
plt.xlabel("Predicted data")
plt.ylabel("Actual data")
plt.show()
```
